```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import shap

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (classification_report, confusion_matrix,
                             roc_auc_score, precision_recall_curve, auc)
from sklearn.datasets import make_classification

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from xgboost import XGBClassifier

# Set style for plots
sns.set(style="whitegrid")

class ChurnPredictor:
    """
    End-to-end pipeline for Customer Churn Prediction.
    Demonstrates handling imbalanced data, hyperparameter tuning, and explainability.
    """

    def __init__(self, random_state=42):
        self.random_state = random_state
        self.model = None
        self.pipeline = None
        self.X_test = None
        self.y_test = None
        self.feature_names = None

    def create_synthetic_data(self, n_samples=10000, n_features=12, weights=[0.85, 0.15]):
        """
        Generates a synthetic Telco-like churn dataset.
        weights=[0.85, 0.15] simulates a 15% churn rate (Imbalanced).
        """
        print(f"Generating synthetic data with {n_samples} samples and {weights[1]*100}% churn rate...")
        X, y = make_classification(
            n_samples=n_samples,
            n_features=n_features,
            n_informative=8,
            n_redundant=2,
            weights=weights,
            random_state=self.random_state
        )

        # Create a DataFrame with business-sounding feature names
        col_names = [
            'MonthlyCharges', 'TotalCharges', 'Tenure', 'SupportCalls',
            'PaymentDelay', 'UsageData', 'ContractLength', 'Age',
            'Noise1', 'Noise2', 'Noise3', 'Noise4'
        ]
        # Adjust col_names length if n_features changes
        col_names = col_names[:n_features]

        df = pd.DataFrame(X, columns=col_names)
        df['Churn'] = y

        print("Data generated successfully.")
        return df

    def build_pipeline(self):
        """
        Constructs the ML pipeline with SMOTE and XGBoost.
        Crucial: SMOTE is inside the pipeline to ensure it only applies to the Fold/Train set during CV.
        """
        # Define model
        xgb = XGBClassifier(
            objective='binary:logistic',
            eval_metric='logloss',
            use_label_encoder=False,
            random_state=self.random_state
        )

        # Define steps: Scale -> SMOTE (Oversample minority) -> Classifier
        # Note: We use imblearn's Pipeline, not sklearn's, to handle the sampler correctly
        self.pipeline = ImbPipeline([
            ('scaler', StandardScaler()),
            ('smote', SMOTE(random_state=self.random_state, sampling_strategy='auto')),
            ('classifier', xgb)
        ])

        return self.pipeline

    def train(self, df, target_col='Churn'):
        """
        Splits data and trains the model using RandomizedSearchCV for hyperparameter optimization.
        """
        print("\n--- Starting Training Process ---")

        # 1. Split Data
        X = df.drop(columns=[target_col])
        y = df[target_col]
        self.feature_names = X.columns.tolist()

        X_train, self.X_test, y_train, self.y_test = train_test_split(
            X, y, test_size=0.2, random_state=self.random_state, stratify=y
        )

        # 2. Define Hyperparameter Grid
        # Interview Tip: Show you know which params matter (depth, learning_rate, scale_pos_weight)
        param_grid = {
            'classifier__n_estimators': [100, 200],
            'classifier__learning_rate': [0.01, 0.1, 0.2],
            'classifier__max_depth': [3, 5, 7],
            'classifier__subsample': [0.8, 1.0],
            'classifier__colsample_bytree': [0.8, 1.0]
        }

        # 3. Randomized Search
        search = RandomizedSearchCV(
            self.build_pipeline(),
            param_distributions=param_grid,
            n_iter=10,
            scoring='roc_auc',
            cv=3,
            verbose=1,
            n_jobs=-1,
            random_state=self.random_state
        )

        print("Tuning hyperparameters...")
        search.fit(X_train, y_train)

        self.model = search.best_estimator_
        print(f"\nBest Params: {search.best_params_}")
        print(f"Best Validation AUC: {search.best_score_:.4f}")

    def evaluate(self):
        """
        Evaluates the model on the held-out test set using business-relevant metrics.
        """
        print("\n--- Evaluation on Test Set ---")
        y_pred = self.model.predict(self.X_test)
        y_probs = self.model.predict_proba(self.X_test)[:, 1]

        # 1. Classification Report
        print("\nClassification Report:")
        print(classification_report(self.y_test, y_pred))

        # 2. ROC-AUC
        roc = roc_auc_score(self.y_test, y_probs)
        print(f"ROC-AUC Score: {roc:.4f}")

        # 3. Confusion Matrix Plot
        plt.figure(figsize=(6, 5))
        cm = confusion_matrix(self.y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        plt.show()

    def explain_model(self):
        """
        Uses SHAP to explain feature importance.
        """
        print("\n--- Model Explainability (SHAP) ---")
        # Extract the classifier step and the preprocessed input
        model_step = self.model.named_steps['classifier']
        scaler_step = self.model.named_steps['scaler']

        # Transform test data using the scaler (skip SMOTE for inference/explanation)
        X_test_scaled = scaler_step.transform(self.X_test)

        # Create explainer
        explainer = shap.TreeExplainer(model_step)
        shap_values = explainer.shap_values(X_test_scaled)

        # Plot
        plt.figure(figsize=(10, 6))
        shap.summary_plot(shap_values, X_test_scaled, feature_names=self.feature_names)

# --- Execution ---
if __name__ == "__main__":
    # Initialize logic
    predictor = ChurnPredictor()

    # 1. Get Data
    df = predictor.create_synthetic_data()

    # 2. Train
    predictor.train(df)

    # 3. Evaluate
    predictor.evaluate()

    # 4. Explain
    predictor.explain_model()
```
